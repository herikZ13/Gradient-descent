# Optimization Gradient descent
This is an optimization algorithm that finds a possible minimum value of a function.

## Objective Function (Sphere Function)


In this assignment,[ the Sphere function](https://www.sfu.ca/~ssurjano/spheref.html) is used to evaluate the optimization algorithm and check the efficiency of this *gradient-descent algorithm*.

**The Sphere Function**

$$
f(x) = \sum_{i=1}^{n} x_i^2
$$

where:
- $x = (x_1, x_2, \dots, x_n)$ is a vector $n$ -dimensional.
- $f(x)$ the sum of the squares of the components $x$.

In this case we use two dimensions

$$
x^2+y^2−1.
$$


##MODULES

```python
#Libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function f(x, y) = x^2 + y^2 - 1
def f(x, y):
    return x**2 + y**2 - 1

# Gradient of the function
def gradient_f(x, y):
    return np.array([2*x, 2*y])
```

## Theory on the Gradient Descent algorithm

In the gradient-based algorithm, we will use the "derivative" of the objective function and minimize it to reach the objective.

Partial derivatives are used for our function objective $f(x, y) = x^2 + y^2 - 1$ :
- $\frac{\partial f}{\partial x } = 2x$
- $\frac{\partial f}{\partial y} = 2y$

**Point Update:**

The formula to update the point at each iteration is:
$$
\mathbf{x_{\text{new}}} \leftarrow \mathbf{x} - \eta \, \nabla f(\mathbf{x})
$$
Where:
- $\mathbf{x}_{\text{new}}$ is the new point after the update.
- $\mathbf{x}$ is the current point.
- $\eta$ is the **learning rate**.
- $\nabla f(\mathbf{x}_{\text{current}})$ is the gradient of the function at the current point.

**Learning Rate:**

*The learning rate $ \eta$ controls the step size taken at each iteration.*


```python
# Function to be optimized
def gradient_descent(x_init, y_init, learning_rate, num_iter):
    points = [(x_init, y_init)]

    for _ in range(num_iter):
        grad = gradient_f(x_init, y_init)
        #point update
        x_init, y_init = np.array([x_init, y_init]) - learning_rate * grad
        points.append((x_init, y_init))

    return np.array(points)
```

The most important thing in the gradient descent algorithm will depend on `x_init` and `y_init` (the initialization points)

**The user will be responsible for modifying these parameters to reach the objective.**
x

```python
# Values ​​that should be modified
x_init = -1.7
y_init = 1.7
learning_rate = 0.1
num_iter = 20
```

*If the initialization,* `learning rate`, *and* `num_iter` *are not correct the algorithm will work badly*

**Note: you must have run the previous line codes**

```python
# Run gradient descent algorithm
points = gradient_descent(x_init, y_init, learning_rate, num_iter)

# show the last result of the iteration
print(f'Last value f(x, y): {f(points[-1, 0], points[-1, 1])}')
```

# Function Extra (Ackley Function)

**What if we use another function for example una function con Many Local Minima**

- [Ackley Function](https://www.sfu.ca/~ssurjano/ackley.html)

$$
f(\mathbf{x}) = -a \exp\left(-b \sqrt{\frac{1}{d} \sum_{i=1}^{d} x_i^2}\right) - \exp\left(\frac{1}{d} \sum_{i=1}^{d} \cos(c x_i)\right) + a + \exp(1)
$$

We use this:

$$f(x,y)=−20⋅exp(−0.2⋅ \sqrt{0.5⋅(x 2 +y 2)})−exp(0.5⋅(cos(2πx)+cos(2πy)))+exp(20)$$

- $x$ and $𝑦$ are the two dimensions.

- 𝑒 is the mathematical constant.

The function produces around the one global minimum and introduces multiple local minima.

### Term First:

$$
-20 \cdot \exp\left(-0.1\cdot \sqrt{0.5 \cdot (x^2 + y^2)}\right)
$$


### Term Second:

$$
-\exp\left(0.5 \cdot (\cos(\pi\cdot x) + \cos(\pi\cdot y))\right)
$$


### Therefore:

$$ term1 + term2 + a + \exp$$



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def modified_ackley(x, y):
    a = 20
    b = 0.1
    c = np.pi

    term1 = -a * np.exp(-b * np.sqrt(0.5 * (x**2 + y**2)))
    term2 = -np.exp(0.5 * (np.cos(c * x) + np.cos(c * y)))

    return term1 + term2 + a + np.e

def modified_ackley_gradient(x, y):
    a = 20
    b = 0.1
    c = np.pi

    sqrt_term = np.sqrt(0.5 * (x**2 + y**2))
    exp_term1 = np.exp(-b * sqrt_term)
    exp_term2 = np.exp(0.5 * (np.cos(c * x) + np.cos(c * y)))

    grad_x = (a * b * x * exp_term1 / sqrt_term) + (c * np.pi * np.sin(c * x) * exp_term2)
    grad_y = (a * b * y * exp_term1 / sqrt_term) + (c * np.pi * np.sin(c * y) * exp_term2)

    return grad_x, grad_y

def gradient_descent(x_start, y_start, learning_rate, num_iterations):
    x, y = x_start, y_start
    points = [(x, y)]

    for _ in range(num_iterations):
        grad_x, grad_y = modified_ackley_gradient(x, y)
        x -= learning_rate * grad_x
        y -= learning_rate * grad_y
        points.append((x, y))

    return np.array(points)
```

```python
x_start, y_start = -1, 1
learning_rate = 0.01
num_iterations = 100
```

Run the code

```python

points = gradient_descent(x_start, y_start, learning_rate, num_iterations)

# Mostrar el valor final
print(f'final point: {points[-1]}')
print(f'Minimun value: {modified_ackley(points[-1, 0], points[-1, 1])}')
```

```python
x = np.linspace(-2, 2)
y = np.linspace(-2, 2)
X, Y = np.meshgrid(x, y)
Z = modified_ackley(X, Y)

# Graficar los resultados
fig = plt.figure(figsize=(14, 5))

# Gráfica 3D
ax = fig.add_subplot(121, projection='3d')
ax.plot([p[0] for p in points], [p[1] for p in points], [modified_ackley(p[0], p[1]) for p in points], 'ro-', label='Puntos en cada iteración')
ax.scatter(x_start, y_start, modified_ackley(x_start, y_start), color='red', s=100, label='Punto de inicio')
ax.scatter(points[-1, 0], points[-1, 1], modified_ackley(points[-1, 0], points[-1, 1]), color='black', s=100, label='Último punto')
ax.plot_surface(X, Y, Z, cmap='plasma', alpha=0.5)
ax.set_title('Resultados 3D')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.legend()

# Gráfica 2D
plt.subplot(122)
contour = plt.contourf(X, Y, Z, levels=200, cmap='plasma', alpha=0.7)
plt.colorbar(contour, label='f(x, y)')
plt.plot([p[0] for p in points], [p[1] for p in points], 'ro-')
plt.plot(x_start, y_start, 'o', markersize=12, color='red')
plt.plot(points[-1, 0], points[-1, 1], 'ko', markersize=12)
plt.title('Resultados 2D')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
